```
from sklearn import datasets
import numpy as np
from torch.utils.data import Dataset, DataLoader
import torch
from torch import nn
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import LocalOutlierFactor
from sklearn.model_selection import train_test_split
import random
from collections import OrderedDict
from datetime import datetime
from torch.nn import CrossEntropyLoss
import torch.optim as optim
from torch.utils.tensorboard import SummaryWriter
torch.cuda.is_available()
```

```
class Circles(Dataset):
    def __init__(self, n_samples, shuffle=True, noise=0.1, random_state=0, factor=.5, feature_list=['x1', 'x2']):

        self.X, self.y = datasets.make_circles(n_samples=n_samples, shuffle=shuffle, noise=noise,
                                               random_state=random_state,
                                               factor=factor)
        sc = StandardScaler()
        self.X = sc.fit_transform(self.X)
        clf1 = LocalOutlierFactor(n_neighbors=2)
        clf1.fit_predict(self.X)
        neg_factor_ = clf1.negative_outlier_factor_
        non_outlier = [idx for idx, val in enumerate(neg_factor_) if val > -2]
        self.X = [self.X[val] for val in non_outlier]
        self.y = [self.y[val] for val in non_outlier]
        clf1 = LocalOutlierFactor(n_neighbors=2)
        clf1.fit_predict(self.X)
        neg_factor_ = clf1.negative_outlier_factor_
        non_outlier = [idx for idx, val in enumerate(neg_factor_) if val > -2]
        self.X = [self.X[val] for val in non_outlier]
        self.y = [self.y[val] for val in non_outlier]
        self.X = np.array(self.X)
        self.data_to_plot = self.X
        self.data_features = np.empty((0, len(self.X)))
        for elem in feature_list:
            if elem == 'x1':
                self.data_features = np.append(self.data_features, [self.X[:,0]], axis=0)
            if elem == 'x2':
                self.data_features = np.append(self.data_features, [self.X[:,1]], axis=0)
            if elem == 'x1^2':
                self.data_features = np.append(self.data_features, [np.square(self.X[:,0])], axis=0)
            if elem == 'x2^2':
                self.data_features = np.append(self.data_features, [np.square(self.X[:,1])], axis=0)
            if elem == 'x1*x2':
                self.data_features = np.append(self.data_features, [np.multiply(self.X[:,0], self.X[:,1])], axis=0)
            if elem == 'sinx1':
                self.data_features = np.append(self.data_features, [np.sin(self.X[:,0])], axis=0)
            if elem == 'sinx2':
                self.data_features = np.append(self.data_features, [np.sin(self.X[:1])], axis=0)
        self.X = np.transpose(self.data_features)
        self.y = np.array(self.y)
        self.X, self.y = self.X.astype(np.float32), self.y.astype(int)
        
    def __len__(self):
        return len(self.X)

    def __getitem__(self, idx):
        return torch.from_numpy(self.X[idx]), torch.from_numpy(np.array(self.y[idx]))

    def plot_data(self):
        plt.scatter(self.data_to_plot[:,0], self.data_to_plot[:,1], c=self.y)
        plt.show()
    



if __name__ == "__main__":
    data = Circles(500)
    data.plot_data()
    print(len(data))
```

```
class SimpleMLP(nn.Module):
    def __init__(self, layer_activations, layer_n_neurons):
        super(SimpleMLP, self).__init__()
        layers = []
        for i in range(len(layer_n_neurons)):
            layers.append(nn.Linear(layer_n_neurons[i][0], layer_n_neurons[i][1]))
            elem = layer_activations[i]
            if elem == 'ReLU':
                layers.append(nn.ReLU())
            elif elem == 'Tanh':
                layers.append(nn.Tanh())
            elif elem == 'Sigmoid':
                layers.append(nn.Sigmoid())
            elif elem == 'None':
                print('No activation for this layer ', i)
            else:
                print('No such activation available')
        self.net = nn.Sequential(*layers)

    def forward(self, x):
        return self.net(x)


if __name__ == "__main__":
   data = Circles(500)
   model = SimpleMLP(['ReLU', 'ReLU'], [(2, 5), (5, 2)])
   print(model)
```

```
class Trainer:
    def __init__(self, model, lr, optimizer=None, criterion=None):
        self.model = model
        self.criterion = CrossEntropyLoss()
        cuda = torch.cuda.is_available()
        self.device = torch.device('cuda:0' if cuda else 'cpu')
        self.experiment_name = datetime.now().strftime("%Y%m%d_%H%M%S")
        self.optimizer = torch.optim.Adam(model.parameters(), lr)
        self.writer = SummaryWriter('runs/' + self.experiment_name)
        # fixme select loss


        # fixme select optimizer


        #fixme use self.device


        #       

        # todo make dirs for logs



    def fit(self, train_dataloader, n_epochs):
        self.model.train()
        for epoch in range(n_epochs):
            print("epoch: ", epoch)
            epoch_loss = 0
            for i, (x_batch, y_batch) in enumerate(train_dataloader):
                if epoch == 0 and i == 0:
                    self.writer.add_graph(self.model, x_batch)
                self.optimizer.zero_grad()
                output = self.model(x_batch)
                loss = self.criterion(output, y_batch)
                loss.backward()
                self.optimizer.step()
                epoch_loss += loss.item()
            print(loss / len(train_dataloader))
            self.writer.add_scalar('Taining loss epoch end', epoch_loss)

    def predict(self, test_dataloader):
        all_outputs = torch.tensor([], dtype=torch.long)
        self.model.eval()
        with torch.no_grad():
            for i, (x_batch, y_batch) in enumerate(test_dataloader):
                output_batch = self.model(x_batch)
                _, predicted = torch.max(output_batch.data, 1)
                all_outputs = torch.cat((all_outputs, predicted), 0)
        return all_outputs

    def predict_proba(self, test_dataloader):
        all_outputs = torch.tensor([], dtype=torch.float32)
        self.model.eval()
        with torch.no_grad():
            for i, (x_batch, y_batch) in enumerate(test_dataloader):
                output_batch = self.model(x_batch)
                all_outputs = torch.cat((all_outputs, output_batch), 0)
        return all_outputs

    def predict_proba_tensor(self, T):
        all_outputs = torch.tensor([], dtype=torch.float32)
        self.model.eval()
        with torch.no_grad():
            output = self.model(T)
        return output



def get_data_from_datasets(train_dataset, test_dataset):
    X_train = train_dataset.data_to_plot.astype(np.float32)
    X_test = test_dataset.data_to_plot.astype(np.float32)
    Y_train = train_dataset.y.astype(int)
    Y_test = test_dataset.y.astype(int)
    return X_train, X_test, Y_train, Y_test


def predict_proba_on_mesh_tensor(clf, xx, yy):
    q = torch.Tensor(np.c_[xx.ravel(), yy.ravel()])
    print(q)
    print(type(q))
    Z = clf.predict_proba_tensor(q)[:, 1]
    Z = Z.reshape(xx.shape)
    return Z
```

```
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import make_moons
from sklearn.neighbors import KNeighborsClassifier


def make_data():
    X, y = make_moons(n_samples = 1000, noise=0.3, random_state=0)
    X = StandardScaler().fit_transform(X)
    X_train, X_test, y_train, y_test = \
        train_test_split(X, y, test_size=.4, random_state=42)
    return X_train, X_test, y_train, y_test


def make_meshgrid(X_train, X_test, h=.02):
    x_min = min(X_train[:, 0].min(), X_test[:, 0].min()) - .5
    x_max = max(X_train[:, 0].max(), X_test[:, 0].max()) + .5
    y_min = min(X_train[:, 1].min(), X_test[:, 1].min()) - .5
    y_max = max(X_train[:, 1].max(), X_test[:, 1].max()) + .5
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))
    return xx, yy


def predict_proba_on_mesh(clf, xx, yy):
    Z = clf.predict_proba(np.c_[xx.ravel(), yy.ravel()])[:, 1]
    Z = Z.reshape(xx.shape)
    return Z


def plot_predictions(xx, yy, Z, X_train=None, X_test=None, y_train=None, y_test=None,
                     figsize=(10, 10),
                     title="predictions",
                     cm=plt.cm.RdBu,
                     cm_bright=ListedColormap(['#FF0000', '#0000FF'])):
    plt.figure(figsize=figsize)
    plt.contourf(xx, yy, Z, cmap=cm, alpha=.8)
    # Plot the training points
    if X_train is not None:
        plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train, cmap=cm_bright,
                    edgecolors='k')
    # Plot the testing points
    if X_test is not None:
        plt.scatter(X_test[:, 0], X_test[:, 1], c=y_test, cmap=cm_bright,
                    edgecolors='k', alpha=0.6)

    plt.xlim(xx.min(), xx.max())
    plt.ylim(yy.min(), yy.max())
    plt.title(title)
    plt.tight_layout()
    plt.show()


if __name__ == "__main__":
    # make data
    X_train, X_test, y_train, y_test = make_data()

    # make mesh
    xx, yy = make_meshgrid(X_train, X_test)

    # make Classifier and fit it
    clf = KNeighborsClassifier(3)
    clf.fit(X_train, y_train)
    Z = predict_proba_on_mesh(clf, xx, yy)

    plot_predictions(xx, yy, Z, X_train=X_train, X_test=X_test)
```
